```python
!pip3 install pytelegrambotapi
import telebot
import os


token = open(os.path.expanduser('~/.mybottoken')).read().strip() #надо создать файл с токеном здесь же - в юпитере
durs_bot = telebot.TeleBot(token)

doc = open('/home/Petriks/СЗ в ССП от Корсаковой Л.А..doc','rb')
doc_0 = open('/home/Petriks/СЗ на Листова Б.П..doc','rb')
doc_1 = open('/home/Petriks/работа с 9-И.docx','rb')
doc_2 = open('/home/Petriks/Отчет по командировке.doc','rb')
doc_3 = open('/home/Petriks/НОМЕНКЛАТУРА 2020.doc','rb')
doc_4 = open('/home/Petriks/ИС Директор ДУРРС 25.03.20.pdf','rb')

entered_password = set()

def check_authenticated(msg):
    if msg.from_user.id in entered_password:
        return True
    else:
        reply = durs_bot.send_message(message.from_user.id, 'Привет,это бот ДУиРРС! =) Назови пароль!')
        def send_text(message):
            txt = message.text
            if txt.lower() == '20 лет рсхб':
                entered_password.add(message.from_user.id)
                durs_bot.send_message(message.from_user.id, 'Добро пожаловать! Задавай вопрос, постараюсь помочь!')
                durs_bot.send_message(message.from_user.id, 'Бот поддерживает следующие команды: \n\t' +
                                       '/Help_me - для вызова подсказки по основным вопросам')

            else:
                durs_bot.send_message(message.from_user.id, 'Это неверный пароль! Узнай пароль и возвращайся =)')

        durs_bot.register_next_step_handler(reply, send_text)
        return False

        
@durs_bot.message_handler(commands=['start'])
def handle_start(message):
    check_authenticated()

    
@durs_bot.message_handler(commands=['Template'])
def handle_template(message):
    if not check_authenticated():
        return

    durs_bot.send_message(message.from_user.id, 'Я нашел для тебя эти шаблоны!')
    durs_bot.send_document(message.chat.id, doc_01)
    durs_bot.send_document(message.chat.id, doc)
    durs_bot.send_document(message.chat.id, doc_2)

    
@durs_bot.message_handler(commands=['Projects'])
def handle_projects(message):
    if not check_authenticated():
        return

    durs_bot.send_message(message.from_user.id, """Смотри, ДУиРРС сейчас ведет следующие проекты\активности:
    1. Внедрение индивидуальных стандартов руководителя
    2. Повышение оперэффективности в 250 ВСП РФ
    3. Активность "Голос РФ"
    Если нужна доп инфо, пиши мне /Question""")

    
@durs_bot.message_handler(commands=['Help_me'])
def handle_help(message):
    if not check_authenticated():
        return

    durs_bot.send_message(message.from_user.id, """Сейчас постараюсь помочь!
    1. если ты не знаешь, как оформить сз, то читай "Вопрос-ответ" по 9-И или пиши мне "Образец служебки"
    2. Если хочешь найти нормативку, то открывай Портал в СМД
    3. Вся информация по проекту 250 ВСП - в разделе Повышение оперэффективности""")
    durs_bot.send_document(message.chat.id, doc_1)

    
@durs_bot.message_handler(commands=['I_standarts'])
def handle_standart(message):
    if not check_authenticated():
        return

    durs_bot.send_message(message.from_user.id, 'Лови стандарты Директора ДУРРС!')
    durs_bot.send_document(message.chat.id, doc_4)

@durs_bot.message_handler(commands=['Question'])
def handle_question(message):
    if not check_authenticated():
        return

    durs_bot.send_message(message.from_user.id, 'Задай мне вопрос, я постараюсь помочь. \n\t Примерные варианты: номенклатура, Образец сз \n\t Если я ничего не найду для тебя в этот раз, то научусь и найду позднее)')

@durs_bot.message_handler(content_types=['text'])
def get_text_messages(message):
    if not check_authenticated():
        return

    if message.text == 'Образец сз':
        durs_bot.send_message(message.from_user.id, 'Лови образцы сз')
        durs_bot.send_document(message.chat.id, doc)
        durs_bot.send_document(message.chat.id, doc_01)
    elif message.text == 'Отчет по командировке':
        durs_bot.send_message(message.from_user.id, 'смотри, я нашел для тебя, как оформить командировку!')
        durs_bot.send_document(message.chat.id, doc_2)

    elif message.text == 'Телефон' or message.text == 'Телефоны':
        durs_bot.send_message(message.from_user.id, 'Любой телефон можно найти на внутреннем портале в разделе Сервис.Но на всякий случай, телефон приемной ДУиРРС 2335 =)')
    elif message.text == 'Стандарты' or message.text == 'Индивидуальные стандарты' or message.text == 'Стандарты руководителя':
        durs_bot.send_message(message.from_user.id, 'Я нашел для тебя стандарты! Если ничего не понятно, звони Наташе доб 1661')
        durs_bot.send_document(message.chat.id, doc_4)
    elif message.text == 'Номер служебки':
        durs_bot.send_message(message.from_user.id, 'Смотри,я нашел для тебя номенклатуру номенклатуру! n\t\ \
        Не забывай в исходящих - при отправке в РФ- указывать номер своего Управления!')
        durs_bot.send_document(message.chat.id, doc_3)
    else:
        durs_bot.send_message(message.from_user.id, 'К сожалению, я тебя не понял =( Я еще не всему научился ')


durs_bot.polling()
```
